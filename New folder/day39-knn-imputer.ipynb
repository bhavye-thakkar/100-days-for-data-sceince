```python
import pandas as pd
import numpy as np
import matplotlib as mlt
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('train.csv', usecols=['Age', 'Pclass', 'Fare', 'Survived'])
df.sample(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().mean()*100
```

```python
x = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
x_train , x_test , y_train , Y_test = train_test_split(x, y, random_state=4, test_size=0.3)
```

```python
x_train.head()
```

```python
knn = KNNImputer(n_neighbors=10, weights='distance')
x_train_knn = knn.fit_transform(x_train)
X_test__knn = knn.transform(x_test)
```

```python
pd.DataFrame(x_train_knn, columns=x_train.columns)
```

```python
lr = LogisticRegression()
lr.fit(x_train_knn, y_train)
y_pred = lr.predict(X_test__knn)
```

```python
accuracy_score(Y_test, y_pred)
```
