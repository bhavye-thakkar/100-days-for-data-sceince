```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split  
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('train.csv', usecols=['Age', 'Fare', 'Survived'])
df.sample(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
x= df.drop(columns=['Survived'])
y = df['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
x_train
```

```python
imputer = SimpleImputer()
```

```python
x_train_trf = imputer.fit_transform(x_train)
x_test_trf = imputer.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.impute import MissingIndicator
```

```python
clf = LogisticRegression()
```

```python
clf.fit(x_train_trf, y_train)
y_pred = clf.predict(x_test_trf)
```

```python
accuracy_score(y_test, y_pred)
```

```python
mi = MissingIndicator()
mi.fit(x_train)
```

```python
mi.features_
```

```python
x_train_missing = mi.transform(x_train)
x_test_missing = mi.transform(x_test)
```

```python
x_train['Age_na'] = x_train_missing
x_test['Age_na'] = x_test_missing
```

```python
x_train
```

```python
imputer = SimpleImputer()
x_train_trf1 = imputer.fit_transform(x_train)
x_test_trf1 = imputer.transform(x_test)
```

```python
clf = LogisticRegression()
clf.fit(x_train_trf1, y_train)
y_pred2 = clf.predict(x_test_trf1)
```

```python
accuracy_score(y_test, y_pred2)
```
